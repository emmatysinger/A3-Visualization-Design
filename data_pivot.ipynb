```python
df = pd.read_csv('data/census2000.csv')

# Pivot the DataFrame
pivoted_df = df.pivot_table(index=['Age', 'Year'], columns='Sex', values='People').reset_index()

# Rename the columns
pivoted_df.columns = ['Age', 'Year', 'Male Population', 'Female Population']

# Save to a new CSV
pivoted_df.to_csv('data/census2000_transformed.csv', index=False)
```

```python
# Function to calculate weighted average
def weighted_average(group):
    total_weight = (group['Age'] * group['People']).sum()
    total_people = group['People'].sum()
    return total_weight / total_people

# Calculate weighted average age for each year
weighted_avg_age_by_year = df.groupby('Year').apply(weighted_average)

print(weighted_avg_age_by_year)
```
